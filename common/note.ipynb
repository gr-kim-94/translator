```python
from transformers import pipeline

translator = pipeline("translation_en_to_fr", model="t5-small") # Helsinki-NLP/opus-mt-en-fr
src = "I like coffee in the morning because it helps me wake up and stay focused."
# result = translator(src, max_length=60)[0]["translation_text"]
# print(result)

results = translator(src, max_length=128)
for r in results:
    print("**", r["translation_text"], "**")
```
